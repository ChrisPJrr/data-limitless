```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://limitlesstcg.com/tournaments/jp?show=50"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
table_rows = soup.select("tr")
```

```python
table_rows
```

```python
for row in table_rows:
    row.unwrap()
```

```python
table_rows[1].get('data-winner')
```

```python
df = pd.DataFrame(columns=['City', 'Date', 'Winner'])

for row in range (1, len(table_rows)):
    aux_row = []
    aux_row.append(table_rows[row].get('data-city'))
    aux_row.append(table_rows[row].get('data-date'))
    aux_row.append(table_rows[row].get('data-winner'))

    df.loc[len(df)] = aux_row

    if row == 42:
        break
```

```python
df
```

```python
df['Winner'].replace('ogerpon raging-bolt', 'raging-bolt ogerpon', inplace=True)
df['Winner'].replace('lugia cinccino', 'lugia archeops', inplace=True)
df['Winner'].replace('ogerpon-cornerstone snorlax', 'snorlax ogerpon-cornerstone', inplace=True)
df['Winner'].replace('roaring-moon squawkabilly', 'roaring-moon', inplace=True)
```

```python
df.groupby('Winner').size().sort_values(ascending=False)
```
